```python
# Diccionario principal
dicts = {
    "CA1d_offsets": CA1d_offsets,
    "PFC_offsets": PFC_offsets,
    "RSC_offsets": RSC_offsets,
    "CA1v_offsets": CA1v_offsets,
}

# Carpeta de salida
output_dir = "npy_points"
os.makedirs(output_dir, exist_ok=True)


# ✅ Función para calcular el punto absoluto
def compute_point(base, offset):
    """Calcula las coordenadas absolutas sumando base + offset."""
    return tuple(np.add(base, offset))


# ✅ Función para guardar archivos .npy con coordenadas y colores
def save_offsets_as_npy(dicts, output_dir, custom_colors):
    """Guarda los offsets en archivos npy: coordenadas y colores."""
    # Coordenadas base (bregma)
    atlas_res = 39
    bregma = (371 * atlas_res, 72 * atlas_res, 266 * atlas_res)

   # Guardar puntos, colores e IDs para cada región
    for region_name, offsets in dicts.items():
        # Calcular coordenadas absolutas
        coords = [compute_point(bregma, offset) for offset in offsets.values()]

        # Asignar colores (estrictamente igual a la cantidad de puntos)

        region_ids = list(offsets.keys())  # Obtener los IDs de los puntos

        # Guardar archivos .npy
        np.save(os.path.join(output_dir, f"{region_name}_coords.npy"), coords)
        np.save(os.path.join(output_dir, f"{region_name}_colors.npy"), custom_colors)
        np.save(os.path.join(output_dir, f"{region_name}_ids.npy"), region_ids)

        print(f"Guardado: {region_name}_coords.npy, {region_name}_colors.npy y {region_name}_ids.npy")


# ✅ Ejecutar la función
save_offsets_as_npy(dicts, output_dir, custom_colors)
```
